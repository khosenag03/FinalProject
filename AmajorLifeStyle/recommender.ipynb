```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import pandas as pd
```

```python
csv = "recommendation.csv"
df = pd.read_csv(csv, encoding="ISO-8859-1")
df.shape
```

```python
df['School ID'] = df['Unnamed: 0']
new_df = df.drop(['Unnamed: 0'],axis=1)

```

```python
new_df = new_df.replace(['Southern','Midwestern', 'Western', 'California', 'Northeastern' ], [0,1,2,3,4])
new_df[['School ID', 'REGION']] = new_df[['School ID', 'REGION']].apply(pd.to_numeric, errors='coerce')
```

```python
new_df.head(10)
```

```python
new_df.loc[:,'EAT-IN'] *= 30
new_df.loc[:,'GAS'] *= 30
new_df
```

```python
edited_df = new_df.drop(['School Name'], axis=1)
edited_df = edited_df.astype('int32')
edited_df.head()
```

```python
edited_df.columns=['STARTING SALARY','MID-CAREER SALARY', '90TH PERCENTILE','MORTGAGE','RENT','ELECTRICITY','GAS','FOOD','PHONE','HLTH & WELL','VET','REGION','SCHOOL-ID']
edited_df.head()


```

```python
edited_df = edited_df.drop(['90TH PERCENTILE'],axis=1)
edited_df.head()
```

```python
edited_df.to_csv("FinalLifestyleDF.csv")
```
