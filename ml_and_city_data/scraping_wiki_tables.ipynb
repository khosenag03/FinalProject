```python
sal_reg = "salaries-by-region.csv"
df = pd.read_csv(sal_reg, encoding="ISO-8859-1")
df
```

```python
new_df = df.replace(abbr, State_list)
new_df.head()
```

```python
edit_df= new_df.replace(' ', '-', regex=True)


city_state_df = edit_df[['State','City']]

city_state_df

city_state_df["state_city"] = city_state_df["State"].map(str) + '-' + city_state_df["City"]

city_state_list = city_state_df['state_city'].tolist()

city_state_list



```

```python
pages = []
for page in city_state_list:
    page_url = f"{base_url}{page}"
    pages.append(page_url)
    print(page_url)
```

```python
pages_df = pd.DataFrame(pages)

pages_df.to_csv('city_state_list.csv')
```
