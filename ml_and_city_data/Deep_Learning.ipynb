```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import pandas as pd
```

```python
csv = "ml.csv"
df = pd.read_csv(csv, encoding="ISO-8859-1")
df.shape
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
southern_df = df.loc[df['REGION'] == 'Southern']
Midwestern_df = df.loc[df['REGION'] == 'Midwestern']
Western_df = df.loc[df['REGION'] == 'Western']
Northeast_df = df.loc[df['REGION'] == 'Northeastern']
Cali_df = df.loc[df['REGION'] == 'California']

avg_southern = pd.DataFrame(southern_df.mean())
avg_Midwestern = pd.DataFrame(Midwestern_df.mean())
avg_west = pd.DataFrame(Western_df.mean())
avg_ne = pd.DataFrame(Northeast_df.mean())
avg_cali = pd.DataFrame(Cali_df.mean())

avg_southern = avg_southern.reset_index()
avg_Midwestern = avg_Midwestern.reset_index()
avg_west = avg_west.reset_index()
avg_ne = avg_ne.reset_index()
avg_cali = avg_cali.reset_index()


merged = pd.merge(avg_southern, avg_Midwestern, on='index')

merged = pd.merge(merged, avg_west, on='index')
merged = pd.merge(merged, avg_ne, on='index')
merged = pd.merge(merged, avg_cali, on='index')

merged.columns = ['AVG', 'SOUTHERN', 'MIDWESTERN', 'WESTERN','NORTHEASTERN', 'CALIFORNIA']

#merged.to_csv('region_avgs.csv')





```

```python
new_df = df.replace(['Southern','Midwestern', 'Western', 'California', 'Northeastern' ], [0,1,2,3,4])
new_df = new_df.drop(['Unnamed: 0', 'Starting Median Salary', 'Mid-Career 90th Percentile Salary'], axis=1)

new_df.head()
```

```python
X = new_df.drop('REGION', axis=1)
y = new_df['REGION']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=8, activation='relu', input_dim=9))
model.add(Dense(units=5, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=500,
    shuffle=True,
    verbose=2
)
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=7, activation='relu', input_dim=9))
deep_model.add(Dense(units=7, activation='relu'))
deep_model.add(Dense(units=5, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=500,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
