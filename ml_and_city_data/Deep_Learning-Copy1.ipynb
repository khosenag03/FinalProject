```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import pandas as pd
```

```python
csv = "LifstyleByRegion2.csv"
df = pd.read_csv(csv, encoding="ISO-8859-1")
df.shape
```

```python
df['Mortgage Payment'] = df['Mortgage Payment'].str.replace(',', '')
df['MORTGAGE'] = df['MORTGAGE'].str.replace(',', '')
df['Starting Median Salary'] = df['Starting Median Salary'].str.replace(',', '')
df['Mid-Career Median Salary'] = df['Mid-Career Median Salary'].str.replace(',', '')
df['Mid-Career 90th Percentile Salary'] = df['Mid-Career 90th Percentile Salary'].str.replace(',', '')
df['Monthly Expense - Mortgage'] = df['Monthly Expense - Mortgage'].str.replace(',', '')
df['Monthly Expense - Rent'] = df['Monthly Expense - Rent'].str.replace(',', '')
df['Min Annual Expense - Mortgage'] = df['Min Annual Expense - Mortgage'].str.replace(',', '')
df['Min Annual Expense - Rent'] = df['Min Annual Expense - Rent'].str.replace(',', '')
df['Adj Annual Expense (30%) - Mortgage'] = df['Adj Annual Expense (30%) - Mortgage'].str.replace(',', '')
df['Adj Annual Expense (30%) - Rent'] = df['Adj Annual Expense (30%) - Rent'].str.replace(',', '')

df.head()
```

```python
df[['Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 90th Percentile Salary', 'MORTGAGE', 'Mortgage Payment','Monthly Expense - Mortgage','Monthly Expense - Rent', 'Min Annual Expense - Mortgage', 'Min Annual Expense - Rent', 'Adj Annual Expense (30%) - Mortgage', 'Adj Annual Expense (30%) - Rent']] \
= df[['Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 90th Percentile Salary', 'MORTGAGE', 'Mortgage Payment', 'Monthly Expense - Mortgage','Monthly Expense - Rent', 'Min Annual Expense - Mortgage', 'Min Annual Expense - Rent', 'Adj Annual Expense (30%) - Mortgage', 'Adj Annual Expense (30%) - Rent']].apply(pd.to_numeric, errors='coerce')

df.dtypes
```

```python
new_df = df.replace(['Southern','Mid-West', 'Western', 'California', 'Northeastern' ], [0,1,2,3,4])

new_df.head()
```

```python
X = new_df.drop('REGION', axis=1)
y = new_df['REGION']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=11))
model.add(Dense(units=5, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=350,
    shuffle=True,
    verbose=2
)
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=11))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=5, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=350,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
