```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
from copy import deepcopy
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
base_url = 'https://www.payscale.com/cost-of-living-calculator/'
```

```python
city_list = ['Illinois-Chicago', 'Ohio-Cincinnati','Ohio-Cleveland', 'Michigan-Detroit', 'Missouri-Kansas-City', \
             'Wisconsin-Milwaukee', 'Minnesota-Minneapolis', 'Missouri-St.-Louis']
count = 0
pages = []
for page in city_list:
    page_url = f"{base_url}{city_list[count]}"
    count = count + 1
    pages.append(page_url)
    print(page_url)
```

```python
some_list = ['Median Home Price','Median Rent','Energy Bill','Phone Bill','Gas','Loaf of Bread', \
             'Gallon of Milk','Carton of Eggs','Bunch of Bananas','Hamburger',"Doctor's Visit",\
             'Dentist Visit','Optometrist Visit','Rx Drug','Veterinary Visit']
```

```python
count = 0
final_df = pd.DataFrame()
for page in pages:
    browser.visit(page)
    city_html = browser.html
    page_soup = bs(city_html, 'html.parser')
    values = []
    for info in page_soup.find_all('div', class_='itemized__item-container'): 
        for v in info.find_all('div', class_='itemized__item-value'):
            value = v.text
            values.append(deepcopy(value))
        
    test_df = pd.DataFrame([{some_list[x]:values[x] for x in range(len(some_list))}]) 
    final_df = final_df.append(deepcopy(test_df))
    
    
```

```python
df = pd.DataFrame(final_df)
df['STATES'] = city_list
```

```python
Mid_west_data = df.set_index('STATES')
Mid_west_data
```

```python
for x in Mid_west_data:
    Mid_west_data[x] = Mid_west_data[x].str.replace("$","")
    Mid_west_data[x] = Mid_west_data[x].str.replace("month","")
    Mid_west_data[x] = Mid_west_data[x].str.replace("/","")
    Mid_west_data[x] = Mid_west_data[x].str.replace("gallon","")
    Mid_west_data[x] = Mid_west_data[x].str.replace(",","")
    Mid_west_data[x] = pd.to_numeric(Mid_west_data[x])
Mid_west_data.head()
```

```python
Mid_west_data['HLTH & WELL'] = Mid_west_data['Dentist Visit'] + Mid_west_data["Doctor's Visit"] + Mid_west_data["Optometrist Visit"] + Mid_west_data["Rx Drug"]
Mid_west_data['EAT-IN'] = Mid_west_data['Bunch of Bananas'] + Mid_west_data['Carton of Eggs'] + Mid_west_data['Gallon of Milk'] + Mid_west_data['Loaf of Bread'] + Mid_west_data['Hamburger']
Mid_west_data = Mid_west_data.drop(['Bunch of Bananas', 'Carton of Eggs', 'Dentist Visit', "Doctor's Visit", 'Gallon of Milk', "Hamburger", "Loaf of Bread", "Optometrist Visit", "Rx Drug"], axis=1)
Mid_west_data
```

```python
Mid_west_data.columns = ['ELECTRICITY', 'GAS','MORTGAGE','RENT','PHONE', 'VET', 'HLTH & WELL','EAT-IN']
Mid_west_data


```

```python
Region = "Mid-West"
Mid_west_data['REGION'] = [Region for _ in range(len(Mid_west_data))] 
Mid_west_data
```

```python
Mid_west_data = Mid_west_data[['MORTGAGE', 'RENT', 'ELECTRICITY','GAS', 'EAT-IN','PHONE','HLTH & WELL', 'VET', 'REGION']]
Mid_west_data.reset_index()
```
