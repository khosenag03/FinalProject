```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
from copy import deepcopy
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
base_url = 'https://www.payscale.com/cost-of-living-calculator/'
```

```python
city_list = ['Massachusetts-Boston', 'New-York-New-York', 'Pennsylvania-Pittsburgh', 'Pennsylvania-Philadelphia', 'Maryland-Baltimore']
count = 0
pages = []
for page in city_list:
    page_url = f"{base_url}{page}"
    pages.append(page_url)
    print(page_url)
```

```python
some_list = ['Median Home Price','Median Rent','Energy Bill','Phone Bill','Gas','Loaf of Bread', \
             'Gallon of Milk','Carton of Eggs','Bunch of Bananas','Hamburger',"Doctor's Visit",\
             'Dentist Visit','Optometrist Visit','Rx Drug','Veterinary Visit']
```

```python
count = 0
final_df = pd.DataFrame()
for page in pages:
    browser.visit(page)
    city_html = browser.html
    page_soup = bs(city_html, 'html.parser')
    values = []
    for info in page_soup.find_all('div', class_='itemized__item-container'): 
        
        for v in info.find_all('div', class_='itemized__item-value'):
            value = v.text
            values.append(deepcopy(value))
        
    test_df = pd.DataFrame([{some_list[x]:values[x] for x in range(len(some_list))}]) 
    print(f"Results from page {page}: {test_df.head()}")
    final_df = final_df.append(deepcopy(test_df))
    
    
```

```python
final_df.reset_index(drop=True)
```

```python
last_df = pd.DataFrame(final_df)
last_df['STATES']= city_list
Northeast_data = last_df.set_index('STATES')
Northeast_data
```

```python
for x in Northeast_data:
   Northeast_data[x] = Northeast_data[x].str.replace("$","")
   Northeast_data[x] = Northeast_data[x].str.replace("month","")
   Northeast_data[x] = Northeast_data[x].str.replace("/","")
   Northeast_data[x] = Northeast_data[x].str.replace("gallon","")
   Northeast_data[x] = Northeast_data[x].str.replace(",","")
   Northeast_data[x] = pd.to_numeric(Northeast_data[x])
Northeast_data.head()
```

```python
region = pd.DataFrame(Northeast_data.mean().round(2))
region_t = region.T
region_t
```

```python
Northeast_data['HLTH & WELL'] = Northeast_data['Dentist Visit'] + Northeast_data["Doctor's Visit"] + Northeast_data["Optometrist Visit"] + Northeast_data["Rx Drug"]
Northeast_data['EAT-IN'] = Northeast_data['Bunch of Bananas'] + Northeast_data['Carton of Eggs'] + Northeast_data['Gallon of Milk'] + Northeast_data['Loaf of Bread'] + Northeast_data['Hamburger']
Northeast_data = Northeast_data.drop(['Bunch of Bananas', 'Carton of Eggs', 'Dentist Visit', "Doctor's Visit", 'Gallon of Milk', "Hamburger", "Loaf of Bread", "Optometrist Visit", "Rx Drug"], axis=1)
Northeast_data
```

```python
Northeast_data.columns = ['ELECTRICITY', 'GAS','MORTGAGE','RENT','PHONE', 'VET', 'HLTH & WELL','EAT-IN']
Northeast_data
```

```python
Region = "Northeast"
Northeast_data['REGION'] = [Region for _ in range(len(Northeast_data))]
Northeast_data
```

```python
Northeast_data = Northeast_data[['MORTGAGE', 'RENT', 'ELECTRICITY','GAS', 'EAT-IN','PHONE','HLTH & WELL', 'VET', 'REGION']]
Northeast_data.reset_index()
```

```python
Northeast_data.to_csv("Northeast.csv")
```
